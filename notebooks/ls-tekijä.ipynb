```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
degrees = [i for i in range(0,91,5)]
degrees = np.array(degrees)
degrees
```

```python
sin_degrees = np.sin(degrees)
sin_degrees
```

```python
#16.8 × sin sg − 0.5, where slope gradient ≥ 0.09
#10.8 × sin sg + 0.03, where slope gradient < 0.09

s = np.where(degrees>=0.09,16.8*degrees-0.5,10.8 * sin_degrees+0.03)
#s = 16.8 * sin_degrees.where(sin_degrees>0.09) - 0.5
s
```

```python
plt.scatter(degrees,s)
plt.show()
```
